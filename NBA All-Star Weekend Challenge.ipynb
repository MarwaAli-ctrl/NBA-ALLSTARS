# What is the average weight of all players who played during this time?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
NBA = pd.read_csv('NBA All Stars 2000-2016 - Sheet1.csv')
```

```python
NBAWT = NBA.groupby('WT').mean()
NBAWT
```

# What team is represented the most during this time?

```python
NBA.groupby('Team').count().sort_values('Year', ascending = False)
#Miami Heat
```

# What draft class is represeted the most during this time?

```python
NBA.groupby('NBA Draft Status').count().sort_values('Year', ascending = False)
```

# Are foriegn players more prevalent in the All-Star Game during this time?
